```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("List of Countries by Sugarcane Production.csv")
df.head()
```

```python
df.tail()
```

```python
df.shape
```

## Data Cleaning

```python
df["Production (Tons)"]=df["Production (Tons)"].str.replace(".","")
df["Production per Person (Kg)"]=df["Production per Person (Kg)"].str.replace(".","").str.replace(",",".")
df["Acreage (Hectare)"]=df["Acreage (Hectare)"].str.replace(".","")
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].str.replace(".","").str.replace(",",".")
```

```python
df.head()
```

```python
df.rename(columns={"Production (Tons)":"Production(Tons)","Production per Person (Kg)":"Production_per_Person(Kg)","Acreage (Hectare)":"Acreage(Hectare)","Yield (Kg / Hectare)":"Yield(Kg/Hectare)"},inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df[df["Acreage(Hectare)"].isnull()]
```

```python
df[df["Continent"]=="Africa"]
```

```python
df=df.dropna().reset_index()
```

```python
df.tail()
```

```python
df.drop(["index","Unnamed: 0"],inplace=True,axis=1)
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df["Production(Tons)"]=df["Production(Tons)"].astype("float")
df["Production_per_Person(Kg)"]=df["Production_per_Person(Kg)"].astype("float")
df["Acreage(Hectare)"]=df["Acreage(Hectare)"].astype("float")
df["Yield(Kg/Hectare)"]=df["Yield(Kg/Hectare)"].astype("float")
```

```python
df.dtypes
```

```python
df.nunique()
```

## Continental Analysis

```python
df_continent=df.groupby("Continent").sum()
df_continent
```

### 1. Analyzing Sugarcane Production by Continent(Which continent produces maximum sugarcane)?

```python
df_continent["Production(Tons)"].sort_values(ascending=False).plot(kind="bar")
```

```python
df_continent["Acreage(Hectare)"].sort_values(ascending=False).plot(kind="bar")
```

### 2. Does the Number of Countries in a Continent Affect Sugarcane Production?  ///No

```python
df["Continent"].value_counts()
```

```python
df_continent["Production(Tons)"].sort_values(ascending=False).plot(kind="bar")
```

```python
df_continent["number_of_countries"]=df.groupby("Continent")["Country"].count()
df_continent
```

```python
continent_names=df_continent.index.to_list()
sns.lineplot(data=df_continent,x="number_of_countries",y="Production(Tons)")
plt.xticks(df_continent["number_of_countries"],continent_names,rotation=90)
plt.show()
```
